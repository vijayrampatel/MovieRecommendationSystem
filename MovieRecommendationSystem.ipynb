```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df.iloc[0].keywords
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df.iloc[0].cast
```

```python
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
movies_df.head()
```

```python
movies_df.iloc[0].crew
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']== "Director":
            L.append(i['name'])
    return L
```

```python
movies_df['crew']= movies_df['crew'].apply(convert_crew)
```

```python
movies_df['spoken_languages']= movies_df['spoken_languages'].apply(convert)
```

```python
movies_df.head()
```

```python
movies_df['overview'] = movies_df['overview'].str.split()
```

```python
movies_df.head()
```

```python
movies_df["cast"] = movies_df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df["crew"] = movies_df["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df["spoken_languages"] = movies_df["spoken_languages"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df["keywords"] = movies_df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies_df["genres"] = movies_df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']= movies_df["cast"]+movies_df["crew"]+movies_df["spoken_languages"]+movies_df["keywords"]+movies_df["genres"]+movies_df["overview"] 
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
